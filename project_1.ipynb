```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using one week's data as a sample to explore first 
df_sample = pd.read_csv("http://web.mta.info/developers/data/nyct/turnstile/turnstile_190907.txt")
```

```python
# all of the data for our chosen 3 month period

# Source: http://web.mta.info/developers/turnstile.html
def get_data(week_nums):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
# weeks: sept 7th to december 7th, 2019    
week_nums = [190907, 190914, 190921, 190928, 191005, 191012, 191019,
            191026, 191102, 191109, 191116, 191123, 191130, 191207]
df = get_data(week_nums)
```

```python
# turning DATE and TIME columns of 'object' entries into a single column
# of type=datetime entries
# (alternative method is to use read_csv parameter: pd.read_csv("url", parse_dates=[["DATE", "TIME"]])

df_sample["DATETIME"] = pd.to_datetime(df_sample["DATE"] + " " + df_sample["TIME"])
```

```python
df_sample["ENTRIES"].describe()
```

```python
# cleaning up data column name "EXITS" which had a bunch of trailing spaces

df_sample.rename(columns={'EXITS                                                               ':"EXITS"}, inplace=True)
```

```python
df_sample
```

```python
# dropping unneeded 'DATE' and 'TIME' columns now that we have 'DATETIME': 

df_sample.drop("DATE", axis=1, inplace=True)
df_sample.drop("TIME", axis=1, inplace=True)
```

```python
df_sample
```

```python
# finding actual entries during each time period: 

real_entries = df_sample.groupby(["C/A", "UNIT", "SCP"]).agg({"ENTRIES": "diff"})
real_entries.head(50)
```

```python
# finding actual exits during each time period: 

real_exits = df_sample.groupby(["C/A", "UNIT", "SCP"]).agg({"EXITS": "diff"})
real_exits.head(50)
```

```python
# adding columns of real entries and exits to the DataFrame 

df_sample["REAL_ENTRIES"] = real_entries["ENTRIES"]
df_sample["REAL_EXITS"] = real_exits["EXITS"]
```

```python
df_sample.head(50)
```

```python
# grouping entries and exits by 4-hour period for each turnstile: 
# doing it in this order to confirm data is changing the way we want

df_sorted1 = df_sample.groupby(["C/A", "UNIT", "SCP", "STATION", "DATETIME"])[["REAL_ENTRIES", "REAL_EXITS"]].max()
df_sorted1
```

```python
# same grouping as above but visualized by station first

df_sorted = df_sample.groupby(["STATION", "C/A", "UNIT", "SCP", "DATETIME"])[["REAL_ENTRIES", "REAL_EXITS"]].max()
df_sorted.head(50)
```

```python
# removing NaN values so they don't look like 0 values 
# (okay to do because always resets at midnight and we don't care about that anyway)

df_sorted.dropna(how="any", inplace=True)
```

```python
# grouping by station and 4-hour period (sum over all turnstiles in a station)

df_sorted_station = df_sorted.groupby(["STATION", "DATETIME"])[["REAL_ENTRIES", "REAL_EXITS"]].sum()
df_sorted_station.head(50)
```

```python
# resetting index to flatten multiindex 

df_sorted_station.reset_index(inplace=True)
```

```python
# looking at irregular events. choosing to keep them because there's a lot of data there, 
# and MTA says: Represent the "REGULAR" scheduled audit event (Normally occurs every 4 hours)
#           1. Audits may occur more that 4 hours due to planning, or troubleshooting activities. 
#           2. Additionally, there may be a "RECOVR AUD" entry: This refers to a missed audit that was recovered.
# the window being a little longer is better to keep than lose the entire 4 hours worth of data

df_sample.loc[df_sample.DESC != "REGULAR"]
```
